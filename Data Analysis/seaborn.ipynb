```python
import seaborn as sns
```

```python
### Basic plotting With Seaborn
tips = sns.load_dataset('tips')
tips
```

```python
## Create a ScatterPlot
import  matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill', y = 'tip', data=tips)
plt.title('Scatter plot of Total bills VS Tips')
plt.show()
```

```python
###LINE Plot
sns.lineplot(x ='size', y = 'total_bill', data=tips)
plt.title('Line Plot for Total Bills Vs Tips')
plt.show()
```

```python
## Categorical Plot 
## Bar Plot 

sns.barplot(x = 'day', y = "total_bill", data=tips)
plt.title(
'bar Plot for day Vs Total_bill'
)

plt.show()
```

```python
## Box Plot
sns.boxplot(x = 'day', y = "total_bill", data=tips)
plt.title(
'Box Plot for day Vs Total_bill'
)

plt.show()
```

```python
## Voilin Plot
sns.violinplot(x = 'day', y = "total_bill", data=tips)
plt.title(
'voiliin Plot for day Vs Total_bill'
)

plt.show()
```

```python
## histogram

sns.histplot(tips['total_bill'], bins=10, kde=True)
```

```python
## KDE plot 
sns.kdeplot(tips['total_bill'], fill = True)
```

```python
##paairplot

sns.pairplot(tips)
```

```python
## HeatMap
corr = tips[['total_bill', 'size', 'tip']].corr()

corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
